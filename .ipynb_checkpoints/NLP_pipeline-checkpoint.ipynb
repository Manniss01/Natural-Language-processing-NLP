#Pre-Processing the raw data from the file 

The following step will be following 
1. remove Punctuation
2. Tokenization
3. remove stop words
4. lemmatize/Stem

```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t', header=None)
data.columns = ['label', 'body_text']

data.head()
```

```python
#Remove Punctuation  from the text file 
import string #importing the libarry for the panctuation

string.punctuation
```

```python
"I like dog" == "i like dog." #it's false because python doesn't understand the meaning of it so to fix this
```

```python
def remove_punct(text): #this function remove the punctuation from the text 
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

data["body_text_clean"] = data['body_text'].apply(lambda x: remove_punct(x))
data.head()
```

## Tokenization 

```python
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens 
data['body_text_tokenized'] = data['body_text_clean'].apply(lambda x: tokenize(x.lower()))

data.head()
```

## Removing stopwords from the dataset 

```python
import nltk

stopwords = nltk.corpus.stopwords.words('english') #list of stopwords 
print(stopwords)
```

```python
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopwords]
    return text

data['body_text_nostop'] = data['body_text_tokenized'].apply(lambda x: remove_stopwords(x))

data.head()
```

## Using STEMMING for cleaning data 

```python
ps = nltk.PorterStemmer()
dir(ps)
```

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

data['body_text_stemmed'] = data['body_text_tokenized'].apply(lambda x: stemming(x))

data.head()
```

```python
ws = nltk.WordNetLemmatizer()
```

```python
dir(ws)
```

```python
#difference between lemmatizing and stemming 

#stemming
print(ps.stem('meanness'))
print(ps.stem('meaning'))
```

```python
print(ps.stem('gooes'))
print(ps.stem('geese'))
```

```python
#lemmatizing
print(ws.lemmatize('meanness'))
print(ws.lemmatize('meaning'))
```

```python
print(ws.lemmatize('goose'))
print(ws.lemmatize('geese'))
```

```python
def lemmatizing(tokenized_text):
    text = [ws.lemmatize(word) for word in tokenized_text]
    return text

data['body_text_lemmatized'] = data['body_text_tokenized'].apply(lambda x: stemming(x))

data.head()
```
